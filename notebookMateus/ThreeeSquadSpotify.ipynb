```python
import pandas as pd
import json
import csv
import numpy as np
from pandas import json_normalize as jnz
```

```python
filePathStreaming = (
            '/content/drive/MyDrive/Projeto-Final/'
            'Data_Mateus/spotify/StreamingHistory0.json'
)
```

```python
def aberturaFileJson(fileJsonPath):
   with  open(fileJsonPath) as fileJson: 
      dictFromJson = json.load(fileJson)  
      return dictFromJson
```

```python
def montaDataFrame(dicionarioData):
    return jnz(dicionarioData)
```

```python
def converterMinutos(x):
  return x / 60000
```

```python
def aplica_nome_mes(mes_num):
   for num in mapeamento_mes:
       if str(mes_num) == str(num):
            return mapeamento_mes[num]
```

```python
mapeamento_mes = {
                  '1':'jan',
                  '2':'fev',
                  '3':'mar',
                  '4':'abr',
                  '5':'maio',
                  '6':'jun',
                  '7':'jul',
                  '8':'ago',
                  '9':'set',
                  '10':'out',
                  '11':'nov',
                  '12':'dez'
  }
 
```

```python
streamingStoriesDict = aberturaFileJson(filePathStreaming)
dfStreaming = montaDataFrame(streamingStoriesDict)
```

```python
dfStreaming.iloc[20:100]
```

```python
dfStreaming.info()
```

```python
#Coverter campo endTime para datatime pd
dfStreaming['endTime'] = pd.to_datetime(dfStreaming['endTime'],format='%Y%m%d %H:%M') 
```

```python
#Separarando data de endTime
dfStreaming['endDate'] = dfStreaming['endTime'].dt.date
```

```python
#separando hora de endTime
dfStreaming['endTime'] = dfStreaming['endTime'].dt.time
```

```python
dfStreaming['trackName'].loc[1654]
```

```python
#Convertendo ms em min
dfStreaming['msPlayed'] = dfStreaming['msPlayed'].apply(converterMinutos)
```

```python
dfStreaming.head()
```

```python
#Convertendo para datetime para manipulação
dfStreaming['endDate'] = pd.to_datetime(dfStreaming['endDate'],format='%Y-%m-%d')
```

```python
#Criando novs coluna com nome do mes
dfStreaming['endMonth']  = dfStreaming['endDate'].dt.month.map(aplica_nome_mes)
```
