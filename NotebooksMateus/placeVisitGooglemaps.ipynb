### Funções para este notebook

```python
#Criação de uma dicionario para guardar valores de cada registro presente no arquivo json.
def cria_dicionario_mapeado():
      dataVisitPlaces = {
                      'placeId':[],
                      'name':[],
                      'address':[],
                      'latitudePlacesVisit': [],
                      'longitudePlacesVisit': [],
                      'startTimestampMs': [],
                      'endTimestampMs': []
    } 
      return dataVisitPlaces
```

```python
def abre_arquivo_json(local_arquivo):
     with open(local_arquivo) as json_arquivo:  
        dict_dados = json.load(json_arquivo)
     json_arquivo.close()
     return dict_dados
```

```python
def busca_dados_local_visitado(dado_dicionario):
   result = dado_dicionario['timelineObjects']
   items = []
   for i,place in enumerate(result):
      if 'placeVisit' in result[i]:
        items.append(result[i])
   return items         
```

```python
def busca_informacao_local(items):
   location  = []
   
   if len(items) > 0:
        for i,item in enumerate(items):
           resul = item['placeVisit']
           if 'location' in resul.keys():
               location.append(resul['location'])
        return location       
   else: 
        return None
```

```python
def transforma_em_minutos(x):
 
   try:
      time_ = datetime.strptime(str(x),'%H:%M:%S.%f').time() 
   except:
      time_ = datetime.strptime(str(x),'%H:%M:%S').time()  
   total_minutos  = abs(time_.hour * 60 + time_.minute + time_.second/60)
                                                                                    
   return  total_minutos


```

```python
def adiciona_dado_localizacao(dados,dicionario_mapeado):
     
       for item in dados:
          if 'placeId' in item.keys():
              dicionario_mapeado['placeId'].append(item['placeId'])
          else:
              dicionario_mapeado['placeId'].append('0')

          if 'name' in item.keys():
              dicionario_mapeado['name'].append(item['name'])
          else:
              dicionario_mapeado['name'].append('0')    
          
          if 'address' in item.keys():
              dicionario_mapeado['address'].append(item['address'])
          else:
              dicionario_mapeado['address'].append('0')

          if 'latitudeE7' in item.keys():
              dicionario_mapeado['latitudePlacesVisit'].append(item['latitudeE7'])
          else:
              dicionario_mapeado['latitudePlacesVisit'].append('0')

          if 'longitudeE7' in item.keys():
              dicionario_mapeado['longitudePlacesVisit'].append(item['longitudeE7'])
          else:
              dicionario_mapeado['longitudePlacesVisit'].append('0')    
       return dicionario_mapeado
        
                  
```

```python
def busca_informacao_duracao(items):
   duracao  = []
   if len(items) > 0:
      for i,item in enumerate(items):
           resul = item['placeVisit']
           if 'duration' in resul.keys():
               duracao.append(resul['duration'])    
      return duracao         
   else: 
        return None
```

```python
def adiciona_dado_duracao(dados,dicionario_mapeado):

       for item in dados:
  
          if 'startTimestampMs' in item.keys():
              dicionario_mapeado['startTimestampMs'].append(item['startTimestampMs'])
          else:
              dicionario_mapeado['startTimestampMs'].append('0')
                  
          if 'endTimestampMs' in item.keys():
              dicionario_mapeado['endTimestampMs'].append(item['endTimestampMs'])
          else:
              dicionario_mapeado['endTimestampMs'].append('0') 
       return dicionario_mapeado          
```

```python
#leitura um a um dos arquivos com a chamdas das funções.
%%time
i = 0
arquivo_json = ''
lista_dataframe = []

while (i < tamanho_lista_nome_arquivo):
   dicionario_mapeado = cria_dicionario_mapeado()
   arquivo_json =f'{local_arquivo}{lista_nome_arquivo[i]}'

   dado_dicionario = abre_arquivo_json(arquivo_json)

   placesVisit = busca_dados_local_visitado(dado_dicionario)
   info_local_visitado =  busca_informacao_local(placesVisit)
   dado_location =  adiciona_dado_localizacao(info_local_visitado,dicionario_mapeado)

   info_duration = busca_informacao_duracao(placesVisit)

   duracao_info = adiciona_dado_duracao(info_duration,dicionario_mapeado)
  
   df = pd.DataFrame(dicionario_mapeado)
   lista_dataframe.append(df)
  
   i+=1   
```
